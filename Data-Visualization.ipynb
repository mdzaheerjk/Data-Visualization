# What is Matplotlib?

- matplotlib is a powerful 2D plotting library in Python.

- It allows you to create publication-quality figures in a variety of formats (PNG, SVG, PDF, etc.).

- Commonly used module: matplotlib.pyplot, often imported as plt.


# Why Use Matplotlib?

- Core plotting engine behind libraries like Seaborn and Pandas plots.

- Works with NumPy and Pandas natively.

- Gives you low-level control of every visual element (useful for fine-tuning).

# Seaborn: High-Level Statistical Data Visualization

Seaborn is a powerful Python visualization library built on top of Matplotlib and integrated with pandas. It simplifies many common tasks like:

- Statistical plots

- Categorical data plots

- Distribution plots

- Color themes

- Multi-variable analysis

```python
import seaborn as sns
```

```python
df_quality=df[(df['Quantity']>0) & (df['Quantity']<100)]
```

```python
df_quality.head()
```

#### Show me how transaction quantities (between 1 and 99) are distributed. Bin them into 10 groups, but also draw a smooth probability curve showing where quantity values tend to cluster.

```python
plt.figure(figsize=(12,6))
sns.histplot(df_quality['Quantity'],bins=10,color='skyblue')
plt.title("Distribution of Revenue")
plt.xlabel('Quantity of Bins')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df_quality['Quantity'],bins=10,kde=True,color='skyblue')
plt.title("Distribution of Revenue")
plt.xlabel('Quantity of Bins')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.kdeplot(df_quality['Quantity'],fill=True,color='green')
plt.title("Revenue KDE Curve")
plt.xlabel('Quantity of Bins')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

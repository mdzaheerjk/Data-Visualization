```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_excel('online_retail_II.xlsx',sheet_name='Year 2009-2010')
df2=pd.read_excel('online_retail_II.xlsx',sheet_name='Year 2010-2011')
```

```python
df1.info()
```

```python
df2.info()
```

```python
list(df1.columns)
```

```python
list(df2.columns)
```

```python
df=pd.concat([df1,df2],ignore_index=True)
```

```python
df.shape
```

```python
print(df.columns)
```

```python
df.head(10)
```

```python
df.info()
```

```python
df=df.dropna(subset='Customer ID')
```

```python
df.info()
```

```python
df.loc[100:200]
```

```python
df=df[~df['Invoice'].astype(str).str.startswith('C')]
```

```python
df.shape
```

```python
df=df[df['Quantity']>0]
```

```python
df.shape
```

```python
df=df[df['Price']>0]
```

```python
df.shape
```

```python
df['Revenue']=df['Quantity']*df['Price']
```

```python
df.info()
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
df.head()
```

```python
df['Month']=df['InvoiceDate'].dt.to_period('M').astype(str)
```

```python
df.head()
```

# What is Matplotlib?

- matplotlib is a powerful 2D plotting library in Python.

- It allows you to create publication-quality figures in a variety of formats (PNG, SVG, PDF, etc.).

- Commonly used module: matplotlib.pyplot, often imported as plt.


# Why Use Matplotlib?

- Core plotting engine behind libraries like Seaborn and Pandas plots.

- Works with NumPy and Pandas natively.

- Gives you low-level control of every visual element (useful for fine-tuning).

```python
df.head()
```

```python
montly_revenue=df.groupby('Month')['Revenue'].sum()
montly_revenue
```

```python
plt.figure(figsize=(12,6))
plt.plot(montly_revenue.index,montly_revenue.values,marker='o')
plt.title("Montly Revenue by Period")
plt.xlabel('Months')
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```
